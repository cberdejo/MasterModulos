```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings(action='ignore') 
%matplotlib inline
```

#### 1.1. Load the 'Boston' dataset from Scikit-Learn and convert it into a DataFrame:

```python
data_url = "http://lib.stat.cmu.edu/datasets/boston"
raw_df = pd.read_csv(data_url, sep="\s+", skiprows=22, header=None)
data = np.hstack([raw_df.values[::2, :], raw_df.values[1::2, :2]])
target = raw_df.values[1::2, 2]
```

```python
header = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT']
```

```python
# The variables.
X = data;
Y = target
Y_reg = Y.reshape(-1, 1)
```

```python
# Transform into a DataFrame for regression target dataset.
df_reg = pd.DataFrame(np.append(X,Y_reg,axis = 1))
df_reg.columns = list(header) + ['PRICE']
```

#### DiscretizaciÃ³n de variable objetivo

```python
max_value = Y.max()
bins = [0, max_value / 3, (2 * max_value) / 3, max_value]  # Dividimos en 3 intervalos
# Create groups grupos "accessible", "medium", "expensive"
df_diag = pd.Series(Y)
df_diag = pd.cut(df_diag, bins=bins, labels=["accessible", "medium", "expensive"], include_lowest=True)
df_diag.value_counts()
```

```python
df_diag
```

```python
# Transform into a DataFrame.
df_class = pd.DataFrame(np.append(X,df_diag.values.reshape(-1, 1),axis = 1))
df_class.columns = list(header) + ['PRICE']
```
